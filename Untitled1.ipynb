```python
# Dealing with missing values

import numpy as np
import pandas as pd

# Example DataFrame
data = {
    'country': ['USA', 'Canada', np.nan, 'USA', np.nan],
    'cast': ['Actor A', np.nan, 'Actor C', 'Actor D', 'Actor E'],
    'director': ['Director A', 'Director B', np.nan, 'Director D', np.nan]
}
df = pd.DataFrame(data)

# Fill missing values for 'country' with the mode
df['country'] = df['country'].fillna(df['country'].mode()[0])

#This retrieves the most common value (mode) in the country column. 
#Using [0] ensures you get the first mode in case there are multiple modes.


# Replace missing values for 'cast' and 'director'
df['cast'].replace(np.nan, 'No Data', inplace=True)
df['director'].replace(np.nan, 'No Data', inplace=True)

print(df)

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Category': ['Action', 'Comedy', 'Drama'],
    'Movie': [30, 20, 25],
    'TV Show': [70, 80, 75]
}

mf_ratio = pd.DataFrame(data)

#This line converts the dictionary data into a pandas DataFrame named mf_ratio. 
#This structure allows for easier data manipulation and analysis.

mf_ratio.set_index('Category', inplace=True)

#Setting Index: This sets the 'Category' column as the index of the DataFrame.
#The inplace=True argument modifies the DataFrame in place, so no new DataFrame is created.

# Create the bar plot
fig, ax = plt.subplots(figsize=(10, 6))
mf_ratio.plot(kind='barh', stacked=True, ax=ax, color=['blue', 'orange'])

# Annotate the bars
for i in range(len(mf_ratio)):
    x_position = mf_ratio['Movie'].iloc[i] + mf_ratio['TV Show'].iloc[i] / 2
    
    #.iloc[i]: This method accesses the value at the integer index i within the 'Movie' column.
    #So, if i = 0, it gets the first row's value in the 'Movie' column.

    
    ax.annotate(
        f"{int(mf_ratio['TV Show'].iloc[i] * 100 / (mf_ratio['Movie'].iloc[i] + mf_ratio['TV Show'].iloc[i]))}%", 
        xy=(x_position, i),
        va='center', ha='center',
        fontsize=20, fontweight='light', fontfamily='serif',
        color='white'
    )
    
    ax.annotate(
        "TV Show", 
        xy=(x_position, i - 0.1),  # Adjusted y-position for clarity
        va='center', ha='center',
        fontsize=12, fontweight='light', fontfamily='serif',
        color='white'
    )

# Set labels and title
ax.set_xlabel('Count')
ax.set_ylabel('Category')
ax.set_title('Movies vs. TV Shows')

# Show the plot
plt.show()
```

# Using .apply() method for various purposes


```python
#Square Values in a Series
import pandas as pd

# Sample DataFrame
df = pd.DataFrame({'numbers': [1, 2, 3, 4, 5]})

# Applying a lambda function to square each number
df['squared'] = df['numbers'].apply(lambda x: x ** 2)

print(df)
```

```python
# Convert Strings to Lowercase
# Sample DataFrame
df = pd.DataFrame({'names': ['Alice', 'Bob', 'CHARLIE', 'David']})

# Applying a function to convert names to lowercase
df['lowercase_names'] = df['names'].apply(lambda x: x.lower())

print(df)
```

```python
# Applying a Custom Function
def classify_number(x):
    if x < 0:
        return 'Negative'
    elif x == 0:
        return 'Zero'
    else:
        return 'Positive'

# Sample DataFrame
df = pd.DataFrame({'numbers': [-1, 0, 1, 5, -3]})

# Applying the custom function
df['classification'] = df['numbers'].apply(classify_number)

print(df)
```

```python
# Applying a Function Across Rows
# Sample DataFrame
df = pd.DataFrame({'first_name': ['John', 'Jane'], 'last_name': ['Doe', 'Smith']})

# Function to combine first and last names
df['full_name'] = df.apply(lambda row: f"{row['first_name']} {row['last_name']}", axis=1)

print(df)
```

```python
# Conditional Logic
def assign_grade(score):
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    elif score >= 60:
        return 'D'
    else:
        return 'F'

# Sample DataFrame
df = pd.DataFrame({'student': ['Alice', 'Bob', 'Charlie'], 'score': [85, 92, 78]})

# Applying the function to assign grades
df['grade'] = df['score'].apply(assign_grade)

print(df)
```
