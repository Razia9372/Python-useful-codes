```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
#allowing the figures to be shown in notebook
```

```python
import numpy as np
x= np.linspace(0, 5, 11)
y= x**2
```

```python
x
```

```python
y
```

```python
# Functional
plt.plot(x,y)
#plt.show --> in order to show figures in environments other that notebooks
```

```python
plt.plot(x,y,'y')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('title')
# giving collor, title and labels just like matlab
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
# Matplotlib object-oriented API methods
# 1. we create a figure object
fig = plt.figure()
# 2. we add axes
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('set')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

```python
fig = plt.figure()

#shifting the middle object to the right

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.15, 0.4, 0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('bigger plot')
axes2.set_title('smaller plot')

#The idea behind oop in matplotlib
```
