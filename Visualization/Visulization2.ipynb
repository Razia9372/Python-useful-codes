```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0, 5, 11)
y= x**2
```

```python
%matplotlib inline
#allowing the figures to be shown in notebook
```

```python
fig = plt.figure()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

# Figure Size and DPI


```python
fig, axes = plt.subplots(nrows= 2, ncols=1, figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
# saving a figure as file with different format
fig.savefig('my_picture.png',dpi=200)
```
