```python
# In this notebook we are going to produce random numbers by coin and dice

import pandas as pd
import numpy as np
```

```python
#Simulating 10 fair coin flips
np.random.seed(0)

def frequency_heads(coin_flip_sequence):
    total_heads = len([head for head in coin_flip_sequence if head == 1])
    #counting the number of heads
    return total_heads / len(coin_flip_sequence)

coin_flips = [np.random.randint(0,2) for _ in range(10)]
freq_heads = frequency_heads(coin_flips)
print(f"Frequency of Heads is {freq_heads}")
```

```python
#Simulating 10 fair coin flips
np.random.seed(0)

def frequency_counts(coin_flip_sequence):
    total_heads = len([head for head in coin_flip_sequence if head == 1])
    #counting the number of heads
    total_tails = len([tail for tail in coin_flip_sequence if tail == 0])
    #counting the number of tails
    A = total_heads / len(coin_flip_sequence)
    B = total_tails / len(coin_flip_sequence)
    return A, B
coin_flips = [np.random.randint(0,2) for _ in range(10)]
freq_count = frequency_counts(coin_flips)
print(f"Frequency of Heads is {freq_count[0]}, and frequency of tails {freq_count[1]}")
```

```python
#Producing random numbers by rolling two dices

num_iterations = 36   # the number of random pairs we want

def frequency_counts(coin_flip_sequence):
    total_ones = len([one for one in coin_flip_sequence if one == 1])
    total_twos = len([two for two in coin_flip_sequence if two == 2])
    total_threes = len([three for three in coin_flip_sequence if three == 3])
    total_fours = len([four for four in coin_flip_sequence if four == 4])
    total_fives = len([five for five in coin_flip_sequence if five == 5])
    total_sixes = len([six for six in coin_flip_sequence if six == 6])
    A = total_ones / len(coin_flip_sequence)
    B = total_twos / len(coin_flip_sequence)
    C = total_threes / len(coin_flip_sequence)
    D = total_fours / len(coin_flip_sequence)
    E = total_fives / len(coin_flip_sequence)
    F = total_sixes / len(coin_flip_sequence)
    return A, B, C, D, E, F

for i in range(num_iterations):
    np.random.seed(i)  # Set the seed to a different value each time
    dice_rolls = [np.random.randint(1, 7) for _ in range(2)]  # Simulate dice rolls
    print(f"Iteration {i+1}: Dice rolls are {dice_rolls}")
    freq_count = frequency_counts(dice_rolls)
    print(f"Frequency of each number is {freq_count} accordingly for 1,2,3,4,5,6")
    print ("__________________________________________________________________________\n")
```
